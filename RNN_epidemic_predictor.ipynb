```python
# A handy helper function for visuals to organize histogram plots 
def draw_histograms(df, n_rows, n_cols, figsize=(21,6), bins=10):
    fig=plt.figure()
    col_names= list(df)
    for i, var_name in enumerate(col_names):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=bins,figsize=figsize, ax=ax)
        ax.set_title(var_name)
    fig.tight_layout()  
    plt.show()
    
```

```python
# Min max scale our features for equal positive scale to prepare for power transformation
from sklearn.preprocessing import MinMaxScaler

scalerM = MinMaxScaler(feature_range=(1,2))
scaled_X = scalerM.fit_transform(df_X)

scaled_X_df = pd.DataFrame(scaled_X, columns=feature_names)
# visualize our distributions after min max transformation
draw_histograms(scaled_X_df, 2 , 7)
```

```python
#  create Box-Cox transformation as our power transformation
from scipy.stats import boxcox

class BoxCoxTransformer():  
    def __init__(self, lmbda=None):
        # If lambda values were passed in during initialization, use those
        self.lmbda = lmbda
       
    def transform(self, X):   
        # If given a 1-dimensional X, reshape it to have 2 dimensions
        if len(X.shape) == 1:
            X = X.values.reshape(-1,1)    
        # Create empty output list
        output = []      
        # For each feature...
        for i in range(X.shape[1]):
            # Transform that feature's data using its corresponding lambda value
            X_transformed = boxcox(X[:, i], self.lmbda[i])          
            # Append to output list
            output.append(X_transformed)     
        # Make output list a Numpy array and transpose it
        output = np.array(output).transpose()       
        return output
    
    def fit(self, X):
        # If given a 1-dimensional X, reshape it to have 2 dimensions
        if len(X.shape) == 1:
            X = X.values.reshape(-1,1)
        # Create empty lambda list
        self.lmbda = []  
        # For each feature...
        for i in range(X.shape[1]):
            # Find the best lambda for transforming that feature
            X_transformed, lmbda = boxcox(X[:, i])    
            # Append that lambda to the lambda list
            self.lmbda.append(lmbda)
        return self
```

```python
# Initialize transformation
box_cox = BoxCoxTransformer()
# Learn and save each of the lambdas into box cox object
box_cox.fit(scaled_X)
# Use learned lambdas to transform data 
boxcox_transformed_features = box_cox.transform(scaled_X)
boxcox_df = pd.DataFrame(boxcox_transformed_features, columns=feature_names)
# visualize our distributions after box cox transformation
draw_histograms(boxcox_df, 2, 7)
```

```python
# We see that after Standardization and  MinMax + Power transformation, the distributions 
# of the features are free of extreme  and look much more like normalize distributions
# print our lambdas
print( box_cox.lmbda )
```

```python
# standardize our box-coxed transformed distribution to set them as same-scale for ease of neural net learning 
scalerX = StandardScaler()
scaled_X_final = scalerX.fit_transform(boxcox_transformed_features)

final_X_df = pd.DataFrame(scaled_X_final, columns=feature_names)

draw_histograms(final_X_df, 2, 7)
```

```python
# keras functions for our model
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Bidirectional, LSTM, GRU
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
from keras.callbacks import EarlyStopping
```

```python
# create and fit the bidirectional GRU network
model = Sequential()
# bidirectional LSTM layer of 14 + 14 = 28 total nodes
model.add(Bidirectional(LSTM(14, return_sequences=True, kernel_initializer='glorot_normal', recurrent_dropout=0.5), \
                                             merge_mode='sum', input_shape=(10, 14)))
# bidirectional GRU layer of 14 + 14 = 28 total nodes
model.add(Bidirectional(GRU(14, kernel_initializer='glorot_normal', recurrent_dropout=0.5), merge_mode='sum'))
model.add(Dense(3, kernel_initializer='glorot_normal'))
model.compile(loss='mse', optimizer='adam', metrics=['mae']) # use mae only for observing learning
```

```python
# Blueprint of our model parameters
print(model.summary())
```

```python
# print our model's architecture
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
# define the checkpoint
import h5py

# path to save model
filepath= "../model-and-weights/checkpoint-best-weights.hdf5"
# checkpointing to save model weights as training
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=3, save_best_only=True, mode='min')
# monitior for loss conergence/divergence
callbacks_list = checkpoint

# train the model, store changes in history variable
history = model.fit(train_X, train_Y, epochs=100, batch_size=32, verbose=2, callbacks=[checkpoint],\
                    validation_split=0.2) # make predictions
```

```python
# plot training / validation curves to check for over/under fitting
plt.plot(history.history['val_loss'])
plt.plot(history.history['mean_absolute_error'])
plt.title('Train vs Validation Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right') 
plt.show()
# we observe overfitting, even with recurrent dropout = 0.5
```

```python
# mean absolute precentage error function MAPE, that returns max error index as well as
def mean_abs_percent_error(y_true, y_pred): 
    # turn to numpy arrays for speed
    y_true, y_pred = np.array(y_true), np.array(y_pred) 
    # Calculate and return the mean absolute percent difference
    all_mapes = np.abs((y_true - y_pred) / y_true) * 100
    mean_mape = np.mean(all_mapes) 
    return all_mapes, mean_mape
```

```python
# for bold prnting
BOLD ='\033[1m'
END = '\033[0m'
```

```python
true = test_Y
naive_mapes = []
print('The Mean Absolute Percentage Error ({}MAPE{}) of the {}naive model\'s predictions{} (mean):'.format(BOLD, END, BOLD, END))
for i, target in enumerate(list(df_test_Y)):
    naive_mean = [true[:, i].mean()] * len(true[:, i])
    _, mean_mape = mean_abs_percent_error(true[:, i], naive_mean)
    naive_mapes.append(mean_mape)
    print('{}MAPE{} for parameter {}{}{} : {}{:.5}{}%'.format(BOLD, END, BOLD, target, END, BOLD, mean_mape, END))
    
print("The {}average MAPE{} for {}naive model{} across the 3 parameters: {}{:.4}{}%".format(BOLD, END,BOLD, END, BOLD, np.mean(naive_mapes), END))
```

```python
# get model preds to calculate error of our model's predictions
preds = final_model.predict(final_test_X) 

# To prevent transformation bias, check the raw predictions and real values to guage model if correct model
raw_preds = scalerY.inverse_transform(preds)
true = test_Y
```

```python
# get MAPE of the models raw  predictions values of alpha, beta, and delta  versus true values
# make a data frame series containing
model_mapes = []

print('The Mean Absolute Percentage Error ({}MAPE{}) of {}Our Model\'s Predictions{}: '.format(BOLD, END, BOLD, END))
for i, target in enumerate(list(df_test_Y)):
    _, mean_mape = mean_abs_percent_error(true[:, i], raw_preds[:, i])
    model_mapes.append(mean_mape)
    print('{}MAPE{} for parameter {}{}{} : {}{:.4}{}%'.format(BOLD, END, BOLD, target, END, BOLD, mean_mape, END))

print("The {}average MAPE{} for {}our model {}across the 3 parameters: {}{:.4}{}%".format(BOLD, END,BOLD, END, BOLD, np.mean(model_mapes), END))
```

## V. Investigating Model Errors

```python
# looking for where and when the model was very wrong 
df_mapes = df_final.copy()
df_mapes.head()

# calculate mape for each parameter of each epidemic and append as column
df_cols = list(df_final) # list of column names
length = len(df_cols) // 2 # half-way point between true and pred columns
step = int(len(df_cols)/2) # symmetrically grab corresponding prediction column
for i in range(length): # for each true column
    # calculate absolute error for each parameter
    new_col_name = 'mape_{}'.format(df_cols[i].split('_')[-1])
    # caculate and append the mape error of the parameter as a column
    df_mapes[new_col_name], _ = mean_abs_percent_error(df_final[df_cols[i]].values, df_final[df_cols[i+step]].values)
```

```python
# create column of the avg mape error for each row/epidemic
df_mapes['mape_avg'] = df_mapes[[col for col in list(df_mapes) if col.startswith("mape")]].mean(axis=1)
df_mapes.head(5)
```

```python
# Interesting to note that our error distributions have skews,
# so it's almost always low error, except in certain cases when it's very wrong
mapes= [col for col in list(df_mapes) if col.startswith("mape")]
draw_histograms(df_mapes[mapes] , 1, len(list(mapes)))
```

```python
# sort dataframe by mape avg error to get the top 3 errors
df_mapes.sort_values(by='mape_avg', ascending=False, inplace=True)
df_mapes.head(5)
```
